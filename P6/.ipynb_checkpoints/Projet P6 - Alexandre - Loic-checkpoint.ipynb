## **3. Analyse et Visualisation**


```python
#3.1 NETTOYAGE DES DONNEES:
data_no_dupes=data.drop_duplicates()
data_no_dupes
```

```python
data_no_dupes.describe()
```

```python
# Vérifions s'il y a des valeurs manquantes:
print(data_no_dupes.isnull().values.any(), data_no_dupes.isnull().sum().sum())
# il n y en a pas.
```

```python
# Les variables non pertinantes: 
# education (redondant avec l education number), 
# le final weight (qui est permet juste d énumérer des memes profils), 
# et la relationship (qui est le role dans la famille)

# Je drope donc ces variables:
efficient_data=data_no_dupes.drop(['education','fnlwgt','relationship'], axis=1)
efficient_data
```

```python
# COMMENCONS PAR UN PETIT PAIR PLOT POUR UNE VISION D ENSEMBLE

sns.pairplot(efficient_data, hue='income')
```

```python
# Voyons plus dans les détails les effets de l'age:
sns.histplot(data=efficient_data, x="age", hue="income", palette=["green","blue"])
# On voit que c est entre 35 et 50 ans qu il y a le plus de revenus au dela des 50k, et que c est de 18 a 35 ans qu il y a le plus de revenus inférieurs a 50k
```

```python
# Voyons les effets du genre:
sns.histplot(data=efficient_data, x="gender", hue="income", palette=["green","blue"])
# on voit qu il y a à peu près la même quantité d'hommes et de femmes qui gagnent moins de 50k, mais il y a 5 fois plus d'hommes qui gagne plus de 50k que de femmes.
# ceci dit, l'étude concerne environ 22k hommes pour seulement 14k femmes
```
