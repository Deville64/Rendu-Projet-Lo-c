```python
#remplacement des labels
X = df.drop('income', axis=1)
y = df[['income']]

X = pd.concat((X.drop('workclass', axis=1), pd.get_dummies(X.workclass, drop_first=True)), axis=1)
X = pd.concat((X.drop('marital-status', axis=1), pd.get_dummies(X['marital-status'], drop_first=True)), axis=1)
X = pd.concat((X.drop('occupation', axis=1), pd.get_dummies(X.occupation, drop_first=True)), axis=1)
X = pd.concat((X.drop('race', axis=1), pd.get_dummies(X.race, drop_first=True)), axis=1)
X = pd.concat((X.drop('gender', axis=1), pd.get_dummies(X.gender, drop_first=True)), axis=1)
X = pd.concat((X.drop('native-country', axis=1), pd.get_dummies(X['native-country'], drop_first=True)), axis=1)

y = pd.concat((y.drop('income', axis=1), pd.get_dummies(y.income, drop_first=True)), axis=1)
```

```python
#entrainement
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

MODELE REGRESSION LOGISTIQUE SANS STANDARDISATION

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear')
lr.fit(X_train, y_train.values.ravel())

# évaluation R2
sc_train = lr.score(X_train, y_train)
sc_test = lr.score(X_test, y_test)

print(f"{sc_train=} et {sc_test=}")

#matrice confusion
from sklearn.metrics import confusion_matrix

y_true = y_test
y_pred = lr.predict(X_test)

pd.DataFrame(confusion_matrix(y_true, y_pred),
             columns=[f'predit {k}' for k in lr.classes_],
             index = [f'vrai {k}' for k in lr.classes_])
```

MODELE REGRESSION LOGISTIQUE AVEC STANDARDSCALER

```python
#standardisation StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

#modele

lr.fit(X_train_sc, y_train.values.ravel())

# évaluation R2
sc_train = lr.score(X_train_sc, y_train)
sc_test = lr.score(X_test_sc, y_test)

print(f"{sc_train=} et {sc_test=}")

#matrice confusion

y_true = y_test
y_pred = lr.predict(X_test_sc)

pd.DataFrame(confusion_matrix(y_true, y_pred),
             columns=[f'predit {k}' for k in lr.classes_],
             index = [f'vrai {k}' for k in lr.classes_])
```

MODELE REGRESSION LOGISTIQUE AVEC MINMAX SCALER

```python
#standardisation MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

mM_scaler = MinMaxScaler()
X_train_mM = mM_scaler.fit_transform(X_train)
X_test_mM = mM_scaler.transform(X_test)

#modele

lr.fit(X_train_mM, y_train.values.ravel())


# évaluation R2
sc_train = lr.score(X_train_mM, y_train)
sc_test = lr.score(X_test_mM, y_test)

print(f"{sc_train=} et {sc_test=}")

#matrice confusion

y_true = y_test
y_pred = lr.predict(X_test_mM)

pd.DataFrame(confusion_matrix(y_true, y_pred),
             columns=[f'predit {k}' for k in lr.classes_],
             index = [f'vrai {k}' for k in lr.classes_])
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score

proba = lr.predict_proba(X_test_sc)[:,1]
fpr, tpr, seuils = roc_curve(y_test, proba, pos_label=1, drop_intermediate=False)
```

```python
# Taux de faux/vrais positifs en fonction des différentes valeurs de seuil
plt.plot(seuils[1:], tpr[1:], label='taux de vrais positifs')
plt.plot(seuils[1:], fpr[1:], label='taux de faux positifs')
plt.legend();
```

```python
# Mesure de l'AUC
score_auc = auc(fpr[1:], tpr[1:])

# Visualisation de la courbe ROC et de l'AUC
fig, ax = plt.subplots()
ax.plot(fpr[1:], tpr[1:], label='auc=%1.5f' %score_auc)
ax.plot([0,1], [0,1], 'k--')
ax.text(0.6, 0.4, "plus mauvais que \nle hasard dans \ncette zone")
ax.set_title("Courbe ROC")
ax.legend();
```
